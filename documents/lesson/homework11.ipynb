```python
import os
import shutil
import subprocess
import datetime

class FileExplorer:
    def __init__(self):
        self.current_directory = os.getcwd()

    def print_menu(self):
        print("1. Создать файл")
        print("2. Удалить файл")
        print("3. Переместить файл")
        print("4. Скопировать файл")
        print("5. Переименовать файл")
        print("6. Создать папку")
        print("7. Удалить папку")
        print("8. Переместить папку")
        print("9. Скопировать файл")
        print("10. Переименовать папку")
        print("11. Контроль версий Git")
        print("0. Выйти")

    def create_file(self):
        file_name = input("Введите имя файла: ")
        now = datetime.datetime.now()
        timestamp = now.strftime("%Y%m%d%H%M%S")
        file_name_with_timestamp = f"{file_name}_{timestamp}"
        file_path = os.path.join(self.current_directory, file_name_with_timestamp)
        try:
            with open(file_path, 'w'):
                pass
            print(f"Файл {file_name_with_timestamp} успешно создан.")
        except OSError as e:
            print(f"Ошибка при создании файла: {e}")

    def delete_file(self):
        file_name = input("Введите имя файла: ")
        file_path = os.path.join(self.current_directory, file_name)
        try:
            os.remove(file_path)
            print(f"Файл {file_name} успешно удален.")
        except OSError as e:
            print(f"Ошибка при удалении файла: {e}")

    def move_file(self):
        file_name = input("Введите имя файла: ")
        source_path = os.path.join(self.current_directory, file_name)
        destination_path = input("Введите путь для перемещения файла: ")
        try:
            shutil.move(source_path, destination_path)
            print(f"Файл {file_name} успешно перемещен.")
        except (OSError, shutil.Error) as e:
            print(f"Ошибка при перемещении файла: {e}")
    def rename_file(self):
        file_name = input("Введите имя файла: ")
        new_file_name = input("Введите новое имя файла: ")
        file_path = os.path.join(self.current_directory, file_name)
        new_file_path = os.path.join(self.current_directory, new_file_name)
        try:
            os.rename(file_path, new_file_path)
            print(f"Файл {file_name} успешно переименован в {new_file_name}.")
        except OSError as e:
            print(f"Ошибка при переименовании файла: {e}")

    def copy_file(self):
        file_name = input("Введите имя файла: ")
        source_path = os.path.join(self.current_directory, file_name)
        destination_path = input("Введите путь для копирования файла: ")
        try:
            shutil.copy2(source_path, destination_path)
            print(f"Файл {file_name} успешно скопирован.")
        except (OSError, shutil.Error) as e:
            print(f"Ошибка при копировании файла: {e}")

    def create_directory(self):
        dir_name = input("Введите имя папки: ")
        now = datetime.datetime.now()
        timestamp = now.strftime("%Y%m%d%H%M%S")
        dir_name_with_timestamp = f"{folder_name}_{timestamp}"
        dir_path = os.path.join(self.current_directory, dir_name_with_timestamp)
        try:
            os.mkdir(dir_path)
            print(f"Папка {dir_name_with_timestamp} успешно создана.")
        except OSError as e:
            print(f"Ошибка при создании папки: {e}")

    def delete_directory(self):
        dir_name = input("Введите имя папки: ")
        dir_path = os.path.join(self.current_directory, dir_name)
        try:
            os.rmdir(dir_path)
            print(f"Папка {dir_name} успешно удалена.")
        except OSError as e:
            print(f"Ошибка при удалении папки: {e}")

    def move_directory(self):
        dir_name = input("Введите имя папки: ")
        source_path = os.path.join(self.current_directory, dir_name)
        destination_path = input("Введите путь для перемещения папки: ")
        try:
            shutil.move(source_path, destination_path)
            print(f"Папка {dir_name} успешно перемещена.")
        except (OSError, shutil.Error) as e:
            print(f"Ошибка при перемещении папки: {e}")
    def rename_directory(self):
        file_name = input("Введите имя файла: ")
        new_file_name = input("Введите новое имя файла: ")
        file_path = os.path.join(self.current_directory, file_name)
        new_file_path = os.path.join(self.current_directory, new_file_name)
        try:
            os.rename(file_path, new_file_path)
            print(f"Файл {file_name} успешно переименован в {new_file_name}.")
        except OSError as e:
            print(f"Ошибка при переименовании файла: {e}")

    def copy_directory(self):
        dir_name = input("Введите имя папки: ")
        source_path = os.path.join(self.current_directory, dir_name)
        destination_path = input("Введите путь для копирования папки: ")
        try:
            shutil.copytree(source_path, destination_path)
            print(f"Папка {dir_name} успешно скопирована.")
        except (OSError, shutil.Error) as e:
            print(f"Ошибка при копировании папки: {e}")

    def git_version_control(self):
        subprocess.run(["git", "--version"])

    def run(self):
        while True:
            self.print_menu()
            choice = input("Выберите действие: ")
            if choice == "1":
                self.create_file()
            elif choice == "2":
                self.delete_file()
            elif choice == "3":
                self.move_file()
            elif choice == "2":
                self.rename_file()    
            elif choice == "5":
                self.copy_file()
            elif choice == "6":
                self.create_directory()
            elif choice == "7":
                self.delete_directory()
            elif choice == "8":
                self.move_directory()
            elif choice == "8":
                self.rename_directory()    
            elif choice == "10":
                self.copy_directory()
            elif choice == "11":
                self.git_version_control()
            elif choice == "0":
                break
            else:
                print("Неверный выбор. Пожалуйста, выберите существующий пункт меню.")

if __name__ == "__main__":
    explorer = FileExplorer()
    explorer.run()
```
